```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# the standard data
df_standard = pd.read_csv('./original data/standard/standard.csv', encoding='ANSI', skiprows=14, nrows=5)
df_standard.columns = ['c', 'A']
```

```python
# get the standard equation between c and A
stats.linregress(df_standard['c'], df_standard['A'])
```

```python
def standard_fit(x):
    return 0.18987056025125854*x+0.01211312410512222
def A_2_c(A):
    return (A-0.01211312410512222)/0.18987056025125854
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
standard_c_points = np.linspace(df_standard['c'].min(), df_standard['c'].max(), 100)
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df_standard['c'], df_standard['A'], '-o')
ax.plot(standard_c_points, [standard_fit(c) for c in standard_c_points])
ax.legend(['scatter plot', 'fitting plot'], fontsize=12)
plt.show()
```

### 25°C-1

```python
# import the data
df_25_1 = pd.read_csv('./original data/25/1.csv', encoding='ANSI', header=None)
df_25_1.columns = ['t', 'A']
```

```python
# clean some data
for i in [0, 1, 3, 5, 6, 14]:
    df_25_1['t'].pop(i)
    df_25_1['A'].pop(i)
```

```python
stats.linregress(df_25_1['t'], [A_2_c(A) for A in df_25_1['A']])
```

```python
def fit_25_1(x):
    return -0.0003103486965785013*x+1.5761224933244098
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
fig, ax = plt.subplots(figsize=(8,6))
t_points = np.linspace(df_25_1['t'].min(), df_25_1['t'].max(), 100)
ax.plot(df_25_1['t'], [A_2_c(A) for A in df_25_1['A']],'-o')
ax.plot(t_points, [fit_25_1(t) for t in t_points])
ax.legend(['scatter plot', 'fitting plot'], fontsize=12)
plt.xlabel('t/s', fontname='Calibri', fontsize=14)
plt.ylabel('c/mM', fontname='Calibri', fontsize=14)
plt.xticks(fontname='Calibri', fontsize=12)
plt.yticks(fontname='Calibri', fontsize=12)
plt.title('c-t plot for 25°C, group #1', fontname='Calibri', fontsize=14)
plt.show()
```
