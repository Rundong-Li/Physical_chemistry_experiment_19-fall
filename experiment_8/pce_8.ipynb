```python
import pandas as pd
import numpy as np
from scipy import stats
from pandas import DataFrame
import matplotlib.pyplot as plt
from matplotlib import ticker
from IPython.display import Latex
```

```python
f_2 = open('C:\\Users/rundong-li/Desktop/物理化学实验/8-蔗糖的转化/original data/2mol_H+.csv', encoding='utf-8')
df_2mol = pd.read_csv(f_2)
```

```python
stats.linregress(df_2mol['t (s)'], df_2mol['ln(αt-αinf)'])
```

```python
def fit_2mol(x):
    return -0.0002729096925246267*x+2.707658508585994
```

```python
f_4 = open('C:\\Users/rundong-li/Desktop/物理化学实验/8-蔗糖的转化/original data/4mol_H+.csv', encoding='utf-8')
df_4mol = pd.read_csv(f_4)
```

```python
stats.linregress(df_4mol['t (s)'], df_4mol['ln(αt-αinf)'])
```

```python
def fit_4mol(x):
    return -0.0009846121685912839*x+2.734999148699817
```

```python
print(f'the reaction order of H+ is {np.log(0.0009846121685912839/0.0002729096925246267)/np.log(2)}')
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df_2mol['t (s)'], df_2mol['αt'], '-o')
ax.set_xlabel('t/s', fontname = 'Calibri', fontsize = 14)
ax.set_ylabel('$α_{t}$', fontname = 'Calibri', fontsize = 14)
plt.xticks(fontsize = 10.5)
plt.yticks(fontsize = 10.5)
plt.title('$α_{t}$—t plot for 2mol/L HCl', fontname = 'Calibri', fontsize = 14)
plt.show()
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
t_points = np.linspace(0, df_2mol['t (s)'].max(), 100)
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df_2mol['t (s)'], df_2mol['ln(αt-αinf)'], '--.')
ax.plot(t_points, [fit_2mol(t) for t in t_points])
ax.legend(['scatter plot', 'linear fitting'], fontsize = 12)
ax.set_xlabel('t/s', fontname = 'Calibri', fontsize = 14)
ax.set_ylabel('ln($α_{t}$-$α_{\infty}$)', fontname = 'Calibri', fontsize = 14)
plt.xticks(fontsize = 10.5)
plt.yticks(fontsize = 10.5)
plt.annotate(s = "y=-0.000273x+2.7077, $R^{2}$=0.997", fontsize =12, xy = (2800, 2.0), xytext = (2400, 2.2), arrowprops={"arrowstyle":"->"})
plt.title('ln($α_{t}$-$α_{\infty}$)—t plot for 2mol/L HCl', fontname = 'Calibri', fontsize = 14)
plt.show()
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df_4mol['t (s)'], df_4mol['αt'], '-o')
ax.set_xlabel('t/s', fontname = 'Calibri', fontsize = 14)
ax.set_ylabel('$α_{t}$', fontname = 'Calibri', fontsize = 14)
plt.xticks(fontsize = 10.5)
plt.yticks(fontsize = 10.5)
plt.title('$α_{t}$—t plot for 4mol/L HCl', fontname = 'Calibri', fontsize = 14)
plt.show()
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
t_points = np.linspace(0, df_4mol['t (s)'].max(), 100)
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df_4mol['t (s)'], df_4mol['ln(αt-αinf)'], '--.')
ax.plot(t_points, [fit_4mol(t) for t in t_points])
ax.legend(['scatter plot', 'linear fitting'], fontsize = 12)
ax.set_xlabel('t/s', fontname = 'Calibri', fontsize = 14)
ax.set_ylabel('ln($α_{t}$-$α_{\infty}$)', fontname = 'Calibri', fontsize = 14)
plt.xticks(fontsize = 10.5)
plt.yticks(fontsize = 10.5)
plt.annotate(s = "y=-0.0009846x+2.735, $R^{2}$=0.995", fontsize = 12, xy = (900, 1.9), xytext = (680, 2.1), arrowprops={"arrowstyle":"->"})
plt.title('ln($α_{t}$-$α_{\infty}$)—t plot for 4mol/L HCl', fontname = 'Calibri', fontsize = 14)
plt.show()
```
